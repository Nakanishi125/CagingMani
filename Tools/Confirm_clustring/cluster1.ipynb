```python
import matplotlib as mpl
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import pandas as pd
import numpy as np
import math
```

```python
Data = pd.read_csv("../../../clust.csv", header=None)
row = len(Data)
print(row)
xx = Data.iloc[0:row,0].values.tolist()
yy = Data.iloc[0:row,1].values.tolist()
zz = Data.iloc[0:row,2].values.tolist()
```

```python
fig = plt.figure()
ax = fig.add_subplot(1,1,1)
ax.set_aspect('equal')

ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_xlim(-300,400)
ax.set_ylim(0, 500)

ax.scatter(xx,yy)
```

```python
fig = plt.figure()
ax = Axes3D(fig, auto_add_to_figure=False)
fig.add_axes(ax)

ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('theta')
ax.set_xlim(-300,300)
ax.set_ylim(0, 400)
ax.set_zlim(0, 360)

ax.scatter(xx, yy, zz)
```
