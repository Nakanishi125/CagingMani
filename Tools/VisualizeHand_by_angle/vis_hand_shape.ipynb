```python
# This program is to visualize the positions of objects and Hands.

# How to use
# To show shape condition, please use getVertex function and input target coordination(x,y,th) as arguments
# To show hand configuration, please make ../../vert.csv by implementing path creating program written in C++
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import math
```

```python
ls = 150
ss = 100
lp = 40
sp = 40
```

```python
def getVertex(x, y, th, verx:list, very:list):
    radAngle = math.radians(th)
    aspect = math.atan2(ls, ss)
    r = math.sqrt(ss*ss + ls*ls)

    ref = [x-0.5*r*math.cos(aspect+radAngle), y-0.5*r*math.sin(aspect+radAngle)]
    p2 = [ref[0] + ss*math.cos(radAngle), ref[1] + ss*math.sin(radAngle)]
    p3 = [p2[0] - lp*math.sin(radAngle), p2[1] + lp*math.cos(radAngle)]
    p4 = [p3[0] - ss*math.cos(radAngle), p3[1] - ss*math.sin(radAngle)]

    verx.append(p4[0])
    very.append(p4[1])
    verx.append(ref[0])
    very.append(ref[1])
    verx.append(p2[0])
    very.append(p2[1])
    verx.append(p3[0])
    very.append(p3[1])
    verx.append(p4[0])
    very.append(p4[1])

    p5 = [p4[0] + sp*math.cos(radAngle), p4[1] + sp*math.sin(radAngle)]
    p6 = [p5[0] - (ls-lp)*math.sin(radAngle), p5[1] + (ls-lp)*math.cos(radAngle)]
    p7 = [p6[0] - sp*math.cos(radAngle), p6[1] - sp*math.sin(radAngle)]

    verx.append(p5[0])
    very.append(p5[1])
    verx.append(p6[0])
    very.append(p6[1])
    verx.append(p7[0])
    very.append(p7[1])
    verx.append(p4[0])
    very.append(p4[1])
 
```

```python
verx = []
very = []
getVertex(230,230,285, verx, very)

fig = plt.figure()
ax = fig.add_subplot(111)
ax.set_aspect('equal')
ax.set_xlim(xmin=-100, xmax=300)
ax.set_ylim(ymin=0, ymax=350)

ax.plot(verx, very)
```
